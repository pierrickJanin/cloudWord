```python
# Python libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import re
from nltk.stem import PorterStemmer
from wordcloud import WordCloud
```

```python
def nuageWord(comment_words):
    wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
    return wordcloud
    
    
```

```python
# Python program to generate WordCloud

# importing all necessary modules
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

# Reads 'Youtube04-Eminem.csv' file
df = pd.read_csv(r"C:/Users/pjani/Documents/parcours iml/projet/P5/data/Youtube04-Eminem.csv", encoding ="latin-1")
df=df.sample(2)
print (df)
comment_words = ''
stopwords = set(STOPWORDS)

# iterate through the csv file
for val in df.CONTENT:
     
    # typecaste each val to string
    val = str(val)
    print("\n val:",val)
 
    # split the value
    tokens = val.split()
    print("\n token:",tokens)
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    
    print("\n token_MIN:",tokens)
    comment_words += " ".join(tokens)+" "
    print("\n comment",comment_words)


wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 

wordCloudA=nuageWord(comment_words)
# plot the WordCloud image					
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordCloudA)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```
